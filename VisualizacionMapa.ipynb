```python
# Función para convertir coordenadas UTM a latitud y longitud
def utm_a_latlon(x, y):
    # Inicializar el transformador de coordenadas
    transformer = Transformer.from_crs('EPSG:25830', 'EPSG:4326', always_xy=True)
    
    # Realizar la transformación
    lon, lat = transformer.transform(x, y)
    return lat, lon
```

```python
import folium
from shapely.geometry import Point, LineString
import pandas as pd

#Función para leer los datos del archivo y preparar los conjuntos de datos
def cargar_datos(filename):
    nodos = []
    lineas = []
    
    with open(filename, 'r') as file:
        for line in file:
            partes = line.strip().split(';')
            inicio = tuple(map(float, partes[0].strip('()').split(', ')))
            fin = tuple(map(float, partes[1].strip('()').split(', ')))
            
            # Convertir coordenadas UTM a latitud y longitud
            inicio_latlon = utm_a_latlon(*inicio)
            fin_latlon = utm_a_latlon(*fin)
            
            # Añadir los puntos convertidos a la lista de nodos
            nodos.extend([inicio_latlon, fin_latlon])
            
            # Añadir la línea convertida a la lista de líneas
            lineas.append(LineString([inicio_latlon, fin_latlon]))
    
    return nodos, lineas

# Función para añadir nodos y líneas a un mapa de folium
def crear_mapa(G):
    # Crear un mapa base. Usamos las coordenadas de un nodo para centrar el mapa
    m = folium.Map(location=[40.4168, -3.7038], zoom_start=10)
    
    # Añadir cada nodo al mapa como un marcador
    for nodo in G.nodes():
        print(nodo)
        folium.Marker([nodo[1], nodo[0]],radius=5,
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.7).add_to(m)
    
    # Añadir cada segmento al mapa como una línea
    for edge in G.edges():
        start_node = edge[1]
        end_node = edge[0]
        line = folium.PolyLine([(start_node[1], start_node[0]), (end_node[1], end_node[0])], color="blue", weight=2.5)
        m.add_child(line)
    return m

# Cargar los datos
# nodos, lineas = cargar_datos(enlaces_ter)  # Reemplaza 'tu_archivo.txt' con el nombre real de tu archivo

# # Crear y mostrar el mapa
# m = crear_mapa(nodos, lineas)

#m.save('mapa_rios.html')  # Guarda el mapa como un archivo HTML
```
